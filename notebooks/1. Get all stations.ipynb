```python
stations = []
for feature in tqdm(rj['features']):
    try:
        stations.append({
            "url": feature['id'],
            "station_type": feature['properties']['@type'],
            "longitude": feature['geometry']['coordinates'][0],
            "latitude": feature['geometry']['coordinates'][1],
            "elevation": feature['properties']['elevation']['value'],
            "elevation_units": feature['properties']['elevation']['unitCode'],
            "station_id": feature['properties']['stationIdentifier'],
            "station_name": feature['properties']['name'],
            "station_timezone": feature['properties']['timeZone'],
            "station_county_url": feature['properties']['county']
        })
    except KeyError:
        print(feature)
```

### Pagination

Single request returns up to 500 stations, but the rest can be obtained by using pagination link

```python
rj['pagination']
```

```python
stations = []

url_station_link = f"{weather_gov_api_url}/stations"

while True:
    response = requests.get(url_station_link, headers=headers)

    if response.status_code != 200:
        print(f"Got status code {response.status_code}")
        break
    
    rj = response.json()

    if len(rj['features']) == 0:
        break
    
    for feature in rj['features']:
        try:
            stations.append({
                "url": feature['id'],
                "station_type": feature['properties']['@type'],
                "longitude": feature['geometry']['coordinates'][0],
                "latitude": feature['geometry']['coordinates'][1],
                "elevation": feature['properties']['elevation']['value'],
                "elevation_units": feature['properties']['elevation']['unitCode'],
                "station_id": feature['properties']['stationIdentifier'],
                "station_name": feature['properties']['name'],
                "station_timezone": feature['properties']['timeZone'],
                "station_county_url": feature['properties']['county']
            })
        except KeyError:
            print(feature)

    url_station_link = rj['pagination']['next']
    sleep(2)
```
